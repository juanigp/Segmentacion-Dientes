```python
import torch, torchvision
import torch.utils.data
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix, plot_confusion_matrix
import json
import os

repo_url = 'https://github.com/juanigp/Segmentacion-Dientes'
!git clone {repo_url}
repo_dir_path = os.path.abspath(os.path.join('.', os.path.basename(repo_url)))
%cd {repo_dir_path}
!git pull
```

```python
#dada una carpeta de jsons con dientes enumerados, hacer un pd dataframe
def get_tooth_type(tooth_number):
  if len(tooth_number) == 2:
    c = tooth_number[1]
  else:
    c = tooth_number 
  if (c == '1') or (c == '2'):
    tooth_type = 3
  elif (c == '3'):
    tooth_type = 2
  elif (c == '4') or (c == '5'):
    tooth_type =1
  else:
    tooth_type = 0
  return int(tooth_type)



def jsons_to_df(jsons_path):
  jsons_list = [f for f in os.listdir(jsons_path) if ".json" in f]
  teeth_list = []
  matriz_orden=[7,6,5,4,3,2,1,0,-1,-1,8,9,10,11,12,13,14,15]
  lista_pares = []
  for json_filename in jsons_list:
    #print(json_filename)
    json_file = json.load( open(os.path.join(jsons_path, json_filename), encoding = 'cp1252' ) )
    height = json_file['imageHeight']
    width  = json_file['imageWidth']
    # separar entre los dientes de arriba y abajo
    superiores=[]
    inferiores=[]
    #print(len(json_file['shapes']))

    for shape in json_file['shapes']:
        if(int(shape['label'])<30):
            superiores.append(shape)
        else:
            inferiores.append(shape)
    #ORDENAR superiores
    superiores_ordenado = [None]*16 
    for diente in superiores:
      ubicacion=int(diente['label'])-11
      ubicacion = matriz_orden[ubicacion]
      superiores_ordenado[ubicacion]=diente


    #ORDENAR inferiores
    inferiores_ordenado = [None]*16 
    for diente in inferiores:
      ubicacion=int(diente['label'])-31
      ubicacion = matriz_orden[ubicacion]
      inferiores_ordenado[ubicacion]=diente
        
  #      for x in inferiores_ordenado[ubicacion]
 #       print(x['label'])
#        print("separador")
    for lista_dientes in (inferiores_ordenado,superiores_ordenado):
      for i in range(0,15):
        for j in range(1,6):
          if((j+i)<15):
            shape=lista_dientes[i]
            if(shape==None):
              continue
            x_min1, x_max1 = shape['points'][0][0], shape['points'][1][0]
            y_min1, y_max1 = shape['points'][0][1], shape['points'][1][1]
            x1 = (( x_min1 + x_max1 ) / 2)/width
            y1 = (( y_min1 + y_max1 ) / 2)/height
            ancho1=abs(x_min1-x_max1)/width
            tipo1=get_tooth_type(shape['label'])
            #print("----")
            #print(shape['label'])
            # alto1=abs(y_min-x_max)/height
            ##--------SEGUNDO DIENTE
            shape=lista_dientes[i+j]
            abajo = 1 if lista_dientes == inferiores_ordenado else 0
            if(shape==None):
              continue
            x_min2, x_max2 = shape['points'][0][0], shape['points'][1][0]
            y_min2, y_max2 = shape['points'][0][1], shape['points'][1][1]
            tipo2=get_tooth_type(shape['label'])
            #print(shape['label'])
            #print(j-1)
            x2 = (( x_min2 + x_max2 ) / 2)/width
            y2 = (( y_min2 + y_max2 ) / 2)/height
            ancho2=abs(x_min2-x_max2)/width
            #alto2=abs(y_min-x_max)/height
            distancia= abs(x1-x2)#**2+(y1-y2)**2

            ####PONE SIEMPRE PRIMERO AL DE LA DERECHA
            if x1<x2:
              posicion=(x_min2+x_max1)/2/width
              lista_pares.append([(j-1),x1,y1,x2,y2,ancho1,ancho2,distancia,posicion,tipo1,tipo2])
            else:
              posicion=(x_min1+x_max2)/2/width
              lista_pares.append([(j-1),x2,y2,x1,y1,ancho2,ancho1,distancia,posicion,tipo2,tipo1])
                                               
 # df = pd.DataFrame(lista_pares, columns=['distancia_dientes','ancho1', 'ancho2', 'distancia_pixel',"esAbajo",'tipo1','tipo2'])
  df = pd.DataFrame(lista_pares, columns=['distancia_dientes',"x1","y1","x2","y2",'ancho1', 'ancho2', 'distancia_pixel',"posicion",'tipo1','tipo2']) 
  return df


#hacer un one hot vector dado el numero de categorias, y la categoria en cuestion
def one_hot_encoder(cat_number, total_cats):
  one_hot = np.zeros(total_cats);
  cat_number=int(cat_number)
  one_hot[cat_number] = 1;
  return one_hot

#dataset para el mlp
class Teeth_number_dataset(torch.utils.data.Dataset):
    def __init__(self, jsons_path):
      self.dataframe = jsons_to_df(jsons_path)

    def __len__(self):
      return len(self.dataframe)

    def __getitem__(self, idx):
      distancia_dientes = self.dataframe.iloc[idx]['distancia_dientes']

      ancho1 = self.dataframe.iloc[idx]['ancho1']
      #alto1  = self.dataframe.iloc[idx]['Alto1']
      ancho2 = self.dataframe.iloc[idx]['ancho2']
      #alto2  = self.dataframe.iloc[idx]['Alto2']
      distancia_pixel = self.dataframe.iloc[idx]['distancia_pixel']
      posicion = self.dataframe.iloc[idx]['posicion']

      #esAbajo = self.dataframe.iloc[idx]['esAbajo']

      tipo1 = self.dataframe.iloc[idx]['tipo1']
      tipo1 = one_hot_encoder(tipo1,4)
      tipo2 = self.dataframe.iloc[idx]['tipo2']
      tipo2 = one_hot_encoder(tipo2,4)

      x1=self.dataframe.iloc[idx]['x1']
      y1=self.dataframe.iloc[idx]['y1']
      x2=self.dataframe.iloc[idx]['x2']
      y2=self.dataframe.iloc[idx]['y2']


      distancia_dientes_one_hot = one_hot_encoder(int(distancia_dientes), 5)


   #   input_vector = torch.cat((torch.tensor((ancho1,ancho2,distancia_pixel,posicion), dtype = float),torch.tensor(tipo1),torch.tensor(tipo2)),0)
      input_vector = torch.cat((torch.tensor((x1,y1,x2,y2), dtype = float),torch.tensor((ancho1,ancho2,distancia_pixel,posicion), dtype = float),torch.tensor(tipo1),torch.tensor(tipo2)),0)     

      
      #input_vector = torch.cat((torch.tensor((ancho1,ancho2), dtype = float),torch.tensor((distancia_pixel,esAbajo), dtype = float)),0)
      return (input_vector, distancia_dientes_one_hot)
      
```

```python
num_epochs = 105
batch_size = 64
learning_rate = 2e-3

input_size = 16#12
output_size = 5
hidden_size = 60


class MLP(nn.Module):
  def __init__(self):
    super(MLP, self).__init__()
    self.input_layer = nn.Linear(input_size, hidden_size)
    self.hidden_1 = nn.Linear(hidden_size, hidden_size)
    self.hidden_2 = nn.Linear(hidden_size, hidden_size)
    self.hidden_3 = nn.Linear(hidden_size, hidden_size)
    self.hidden_4 = nn.Linear(hidden_size, hidden_size)


    self.output_layer = nn.Linear(hidden_size, output_size) 
        
  def forward(self, x):
    x = F.relu(self.input_layer(x))
    x = F.relu(self.hidden_1(x))
    x = F.relu(self.hidden_2(x))
    x = F.relu(self.hidden_3(x))
    x = F.relu(self.hidden_4(x))

    x = self.output_layer(x)
    return x

train_dataset = Teeth_number_dataset("/content/DientesMask/data/old_jsons/jsons_dientes_enumerados/train")
test_dataset = Teeth_number_dataset("/content/DientesMask/data/old_jsons/jsons_dientes_enumerados/test")
train_dataloader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_dataloader = torch.utils.data.DataLoader(test_dataset, batch_size=batch_size, shuffle=True)
```

```python
model = MLP()
optimizer = torch.optim.Adam(params=model.parameters(), lr=learning_rate)
criterion = nn.CrossEntropyLoss()
model.cuda()
model.double()

train_loss_avg = []
test_loss_avg = []

for epoch in range(num_epochs):
  model.train()
  train_loss_avg.append(0)
  num_batches = 0
  for i, (input_vector,output_vector) in enumerate(train_dataloader):
    input_vector = input_vector.cuda()
    output_vector = output_vector.cuda()
    predictions = model(input_vector)
    loss = criterion(predictions, torch.max(output_vector, 1)[1])
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    train_loss_avg[-1] += loss.item()
    num_batches += 1

  train_loss_avg[-1] /= num_batches
  print('Epoch [%d / %d] average train loss: %f' % (epoch+1, num_epochs, train_loss_avg[-1]))

  model.eval()
  with torch.no_grad():
    test_loss_avg.append(0)
    num_batches = 0    
    for (input_vector,output_vector) in test_dataloader:
      input_vector = input_vector.cuda()
      output_vector = output_vector.cuda()
      predictions = model(input_vector)
      loss = criterion(predictions, torch.max(output_vector, 1)[1])
      test_loss_avg[-1] += loss.item()
      num_batches += 1

    test_loss_avg[-1] /= num_batches
    print('Epoch [%d / %d] average test loss: %f' % (epoch+1, num_epochs, test_loss_avg[-1]))

```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
plt.plot(train_loss_avg)
plt.plot(test_loss_avg)
plt.legend(('Training loss', 'Validation loss'))
plt.xlabel('Epochs')
plt.ylabel('Cross-entropy loss')
plt.show()
```

```python
test_loss_avg = 0
num_incorrect = 0
num_instances = 0
cm = np.zeros((32,32))
model.eval()
total_predicted_labels = torch.tensor((), dtype = int).cuda()
total_ground_truth_labels = torch.tensor((), dtype = int).cuda()
with torch.no_grad():
  for (input_vector,output_vector) in test_dataloader:
    input_vector = input_vector.cuda()
    output_vector = output_vector.cuda()
    predictions = model(input_vector)
    predicted_labels = torch.max(predictions, 1)[1]
    ground_truth_labels = torch.max(output_vector, 1)[1]
    total_predicted_labels = torch.cat((total_predicted_labels, predicted_labels),0)
    total_ground_truth_labels = torch.cat((total_ground_truth_labels, ground_truth_labels),0)
    num_incorrect += (predicted_labels != ground_truth_labels).sum().item()
    loss = criterion(predictions, torch.max(output_vector, 1)[1])
    test_loss_avg += loss.item()
    num_batches += 1
    num_instances += input_vector.size(0)



test_loss_avg /= num_batches
print('average loss: %f' % (test_loss_avg))
print('accuracy: %f%%' % ( (num_instances-num_incorrect)/num_instances ) )

```
